# OpenCV Sensor Dust Image Inpainting

```python
#Import libraries
import cv2
from matplotlib import pyplot as plt
```

```python
#Open image and inpainting mask
image = cv2.imread("test_dust.png")
mask_in = cv2.imread("test_dust_mask.png")
mask = cv2.cvtColor(mask_in, cv2.COLOR_BGR2GRAY)
```

```python
plt_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt_mask = cv2.cvtColor(mask_in, cv2.COLOR_BGR2RGB)

plt.subplot(121),plt.imshow(plt_image)
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(plt_mask)
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
#Set inpainting options
radius = 10
flags = cv2.INPAINT_TELEA #or use cv2.INPAINT_NS
```

```python
#Perform inpainting
output = cv2.inpaint(image, mask, radius, flags=flags)
```

```python
#Show and save image
cv2.imwrite('test_dust_inpainting.png', output)

plt_out = cv2.cvtColor(output, cv2.COLOR_BGR2RGB)
plt.subplot(),plt.imshow(plt_out)
plt.title('Inpainting Results'), plt.xticks([]), plt.yticks([])
plt.show()
```
